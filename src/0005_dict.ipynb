```python
# dictionaries or dict are my favorite python object type
# it is also heavily used interinally inside python code

# a dict is a key:value pair.  
# key need to be unique and value can be any object type

# empty dict
my_dict = {}

# lets add some things to this dict:
my_dict = {'Dan': 10, 'Chris': 5}
my_dict.update({'Steve': 44})

# dict support len
print (len(my_dict))
print ()

# how to loop through and print a dict
for k,v in my_dict.items():
    print(k, v)
print ()

# lets calc a new "age" for our dict
for k,v in my_dict.items():
    print("my name is: %s and my age will be %s in 5 years" % (k, str(v+5)))
```

```python
# many times you will have to check to see if a key is already in a dict:
# lets add some things to this dict:
my_dict = {'Dan': 10, 'Chris': 5}
my_dict.update({'Steve': 44})


if 'Dan' in my_dict:
    print ('found a key of Dan')
```

```python
# many times I need dictionaries to keep track of something I'm counting
# python support defaultdict types where you can set the default data type.

# so for example lets make a default dict to keep track of "stars" a movie
# are rated at.  Each time we can do += value.  This way you don't have to check
# if you alrady have a key set like in the preivous example.  You can simple do +=

from collections import defaultdict 

my_dict = defaultdict(int) 

my_dict['4star'] += 3
my_dict['3star'] += 1
my_dict['2star'] += 2
my_dict['4star'] += 1

for k,v in my_dict.items():
    print (k,v)
```

```python
# also as of python 3.6 dictionaries guarantee order.
my_dict = {'Dan': 10, 'Chris': 5}
my_dict.update({'Steve': 44})

# how to loop through and print a dict
for k,v in my_dict.items():
    print(k, v)
print ()

for k,v in my_dict.items():
    print(k, v)
print ()
```

```python
# removing an item
my_dict = {'Dan': 10, 'Chris': 5}
my_dict.update({'Steve': 44})

for k,v in my_dict.items():
    print(k, v)
print ()

my_dict.pop('Dan')

for k,v in my_dict.items():
    print(k, v)
print ()
```

```python
# copy a dict
my_dict = {'Dan': 10, 'Chris': 5}
my_dict.update({'Steve': 44})

my_copy = my_dict.copy()
print(my_copy)
```

```python
# destroys a dictionary and removes from ram
my_dict = {'Dan': 10, 'Chris': 5}
my_dict.update({'Steve': 44})

del my_dict
print ("deleted my_dict var.")
```

```python
# adding another value to a dict
thisdict = {
  "ford": "red",
  "chevy": "beige",
  "toyota": "silver"
}
thisdict["subaru"] = "black"
for k,v in thisdict.items():
    print(k, v)
print ()
```

```python
# dict also support nested objects
# I try to avoid this as much as possible but sometimes you cannot avoid it
child1 = {
  "name" : "Emily",
  "year" : 2004
}
child2 = {
  "name" : "Tobias",
  "year" : 2007
}
child3 = {
  "name" : "Linus",
  "year" : 2011
}

myfamily = {
  "child1" : child1,
  "child2" : child2,
  "child3" : child3
}

print (myfamily)
```
