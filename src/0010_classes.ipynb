```python
# data classes also support defaults which is nice so you don't even have
# to set the value each time.  You can also override a param as well

@dataclass
class Cars:
    name: str
    color: str
    cost: float = 20000
    tax: float = .03
    

    def total_cost(self):
        new_value = round(self.cost + (self.cost * self.tax), 2)
        print(f'The cost with taxes for the {self.name} car will be ${new_value}.')

if __name__ == "__main__":
    p1 = Cars('camery', 'red')
    p1.total_cost()
    
    p1 = Cars('camero', 'red', cost=30000)
    p1.total_cost()
```

```python
# the reason to data classes is because it does save space in code 
# here are 2 classes that are identical 

import sys 

try:
    from dataclasses import dataclass
except ImportError:
    msg1 = "sorry  your version of python is not high enough."
    print(msg1)
    print("python3.7 or higher is reqiured.")
    sys.exit(1)

    
@dataclass 
class Cars1:
    color: str
        
    def output_color(self):
        print(f'The cost with taxes for the {self.color}.')


class Cars2:
    def __init__(self, color):
        self.color = color
        
    def output_color(self):
        print(f'The cost with taxes for the {self.color}.')

if __name__ == "__main__":
    p1 = Cars1('black')
    p1.output_color()
    
    p2 = Cars2('black')
    p2.output_color()
    
    # now lets check the size of both objects
    print (f'The size of p1 is {sys.getsizeof(p1)}.')
    print (f'The size of p2 is {sys.getsizeof(p2)}.')
```

```python
# we can take this one step further to save on memory using something called python "slots"
# more information can be found here: https://realpython.com/python-data-classes/
# be careful beause this is on the frindge feature and this explains why:
# https://github.com/ericvsmith/dataclasses/issues/28

# but you can see from the below output the dataclass does take up less space
# this could be signficiate as you add more attributes and/or allot more class objects

import sys 

try:
    from dataclasses import dataclass
except ImportError:
    msg1 = "sorry  your version of python is not high enough."
    print(msg1)
    print("python3.7 or higher is reqiured.")
    sys.exit(1)

@dataclass 
class Cars1:
    __slots__ = ['color']
    color: str
        
    def output_color(self):
        print(f'The cost with taxes for the {self.color}.')


class Cars2:
    def __init__(self, color):
        self.color = color
        
    def output_color(self):
        print(f'The cost with taxes for the {self.color}.')

if __name__ == "__main__":
    p1 = Cars1('black')
    p1.output_color()
    
    p2 = Cars2('black')
    p2.output_color()
    
 # now lets check the size of both objects
    print (f'The size of p1 is {sys.getsizeof(p1)}.')
    print (f'The size of p2 is {sys.getsizeof(p2)}.')
```

```python
# simple example of sub classesing a dataclass

import sys 

try:
    from dataclasses import dataclass
except ImportError:
    msg1 = "sorry  your version of python is not high enough."
    print(msg1)
    print("python3.7 or higher is reqiured.")
    sys.exit(1)

@dataclass 
class Cars1:
    __slots__ = ['color', 'year', 'cost', 'tax']
    color: str
    year: str
    cost: float
    tax: float
        
    def output_color(self):
        print(f'The cost with taxes for the {self.color}.')
        
@dataclass
class Factory(Cars1):
    __slots__ = ['name', 'tax']
    name: str
        
    def get_report(self):
        print (f'The locations is {self.name}. The cost is: {self.cost}')
        print (f'The taxes for the car will be {self.tax * self.cost}')
        
if __name__ == "__main__":
    p1 = Factory(name='Chicago', color='black', year='2020', cost=20000, tax=.2)
    p1.get_report()
    p1.output_color()
```
